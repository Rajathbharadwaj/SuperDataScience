```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(12,9))
sns.boxplot(x='State',y='Profit',data=df)
```

```python
plt.figure(figsize=(12,9))
sns.barplot(x='Profit',y='State',data=df)
```

```python
plt.figure(figsize=(12,9))
plt.scatter('Profit','R&D Spend',data=df)
plt.title('Profit VS R&D Spend')
plt.xlabel('Profit')
plt.ylabel('R&D Spend')
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.barplot('Profit','R&D Spend',data=df)
plt.xticks(rotation=46)
```

```python
df.corr()['Profit']
```

```python
plt.figure(figsize=(12,9))
sns.barplot('Profit','Marketing Spend',data=df)
plt.xticks(rotation=90)
```
